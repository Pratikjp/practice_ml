```
#import libraries
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
df = pd.read_csv('./water-treatment (1).data',header=None)
#df.shape
df.head(5)
```

```
#add column names
df.columns =['Day','Q-E','ZN-E','PH-E','DBO-E','DQO-E','SS-E','SSV-E','SED-E','COND-E','PH-P','DBO-P','SS-P','SSV-P','SED-P','COND-P','PH-D','DBO-D','DQO-D','SS-D','SSV-D','SED-D',
            'COND-D','PH-S','DBO-S','DQO-S','SS-S','SSV-S','SED-S','COND-S','RD-DBO-P','RD-SS-P','RD-SED-P','RD-DBO-S','RD-DQO-S','RD-DBO-G','RD-DQO-G','RD-SS-G','RD-SED-G'] 
#df.head(20)
df.tail(5)
```

```
df = df.replace('?',np.nan)
df.tail(5)
```

```
df.dtypes
```

```
#convert object columns to float 
#df = df.infer_objects()
convert_columns_dtype = {'Q-E': float, 'ZN-E': float,'DBO-E':float,'DQO-E':float,'SS-E':float,'SSV-E':float,'SED-E':float,
                         'DBO-P':float,'SSV-P':float,'SED-P':float,
                         'DBO-D':float,'DQO-D':float,'SS-D':float,'SSV-D':float,'SED-D':float,
                         'PH-S':float,'DBO-S':float,'DQO-S':float,'SS-S':float,'SSV-S':float,'SED-S':float,'COND-S':float,
                         'RD-DBO-P':float,'RD-SS-P':float,'RD-SED-P':float,
                         'RD-DBO-S':float,'RD-DQO-S':float,
                         'RD-DBO-G':float,'RD-DQO-G':float,'RD-SS-G':float,'RD-SED-G':float}
df = df.astype(convert_columns_dtype)
df.dtypes
```

```
#check for null values
df.apply(lambda x:sum(x.isnull()))
```

```
#replace null values with mean
df_new = df.iloc[:,1:]
df_new = df_new.apply(lambda x: x.fillna(x.mean()),axis=0)
df_new.tail()
```

```
#convert columns to native format
convert_columns_dtype_ftoi = {'Q-E': int,'DBO-E':int,'DQO-E':int,'SS-E':int,
                         'DBO-P':int,
                         'DBO-D':int,'DQO-D':int,'SS-D':int,
                         'DBO-S':int,'DQO-S':int,'SS-S':int,'COND-S':int}
df_new = df_new.astype(convert_columns_dtype_ftoi)
df_new.dtypes
df_new.tail()
```

```
#data distribution
import seaborn as sb
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2,4,figsize=(30,20))
fig.subplots_adjust(hspace=0.1, wspace=0.2)

df_new[['PH-E','PH-P','PH-D']].plot.hist(alpha=1.0,ax=ax[0,0])
df_new[['DBO-E','DBO-P','DBO-D','RD-DBO-P','RD-DBO-S','RD-DBO-G']].plot.hist(alpha=1.0,ax=ax[0,1])
df_new[['DQO-E','DQO-D','RD-DQO-S','RD-DQO-G']].plot.hist(alpha=1.0,ax=ax[0,2])
df_new[['SS-E','SS-P','SS-D','RD-SS-P','RD-SS-G']].plot.hist(alpha=1.0,ax=ax[0,3])
df_new[['SSV-E','SSV-P','SSV-D']].plot.hist(alpha=1.0,ax=ax[1,0])
df_new[['SED-E','SED-P','SED-D','RD-SED-P','RD-SED-G']].plot.hist(alpha=1.0,ax=ax[1,1])
df_new[['COND-E','COND-P','COND-D']].plot.hist(alpha=1.0,ax=ax[1,2])
fig.delaxes(ax[1,3])
```

```
#correlation within data
fig, ax = plt.subplots(figsize=(15,15))
sb.heatmap(df_new.corr(),linewidths=2,cmap="viridis",ax=ax)
```

```
#apply PCA to reduce dimensions
import sklearn as sk
from sklearn.decomposition import PCA

scaler= sk.preprocessing.MinMaxScaler()
scaled_dataset = scaler.fit_transform(df_new)

pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_dataset)
plt.scatter(pca_result[:,0],pca_result[:,1],s=50)
```

```
#find number of clusters for kmeans
from sklearn.cluster import KMeans
wcss = []
for i in range(1,10):
  kmeans=KMeans(n_clusters=i, init='k-means++',random_state=0)
  kmeans.fit(pca_result)
  wcss.append(kmeans.inertia_)
#elbow method to find out no of clusters
plt.title("Elbow method")
plt.plot(range(1,10),wcss)
plt.xlabel('No of clusters')
plt.ylabel('wcss')
plt.show()
```

```
kmeans=KMeans(n_clusters=7,init='k-means++',random_state=0)
kmeans_cluster_labels=kmeans.fit_predict(pca_result)

#validate kmeans
from sklearn.metrics import silhouette_score
kmeans_silhouette_avg = silhouette_score(pca_result, kmeans_cluster_labels)
print("The average silhouette_score for kmeans is :" , kmeans_silhouette_avg)
```

```
def plot_clusters(pca_result,cluster_labels,n_clusters):
    colors = ['red','blue','green','yellow','brown','grey','cyan','pink'] 
    plt.figure(figsize=(15,15))
    for i in range(0,n_clusters):
      plt.scatter(pca_result[cluster_labels==i,0],pca_result[cluster_labels==i,1],s=50,c=colors[i], label = "Cluster {}".format(i))
    plt.legend()
    plt.show()

plot_clusters(pca_result,kmeans_cluster_labels,7)
```

```
#explore outcome
k_means_result = df
k_means_result['Cluster'] = kmeans_cluster_labels
#k_means_result.head(20)
#sample cluster values 
print("Cluster 1 : ",np.array(k_means_result[k_means_result['Cluster']==0]['Day'].values))
print("Cluster 4 : ",np.array(k_means_result[k_means_result['Cluster']==3]['Day'].values))
print("Cluster 7 : ",np.array(k_means_result[k_means_result['Cluster']==6]['Day'].values))
```

```
#hierarchical clustering
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch 

#find out no of clusters for Agglomerative Clustering
fig, ax = plt.subplots(figsize=(50,20))
dendogram = sch.dendrogram(sch.linkage(pca_result,'single'),ax=ax)
#number of clusters reading from dendrogram  = 8
```

```
#Apply Agglomerative Clustering
hierarchy = AgglomerativeClustering(n_clusters=8)
hierarchy_cluster_labels=hierarchy.fit_predict(pca_result)
```

```
#plot the results
plot_clusters(pca_result,hierarchy_cluster_labels,8)
```
