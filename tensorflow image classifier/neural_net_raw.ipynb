```
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data 
mnist = input_data.read_data_sets('./data/',one_hot=True)
```

```
#parameters
learning_rate = 0.001
batch_size = 20000
epochs = 1000

#network parameters
n_inputs = 784
n_hidden1 = 256
n_hidden2 = 128
n_out = 10

#graph 
X = tf.placeholder("float",[None,n_inputs])
weights ={
    'w1': tf.Variable(tf.random_normal([n_inputs,n_hidden1],mean=0.5,stddev=0.5)),
    'w2': tf.Variable(tf.random_normal([n_hidden1,n_hidden2],mean=0.5,stddev=0.5)),
    'w3': tf.Variable(tf.random_normal([n_hidden2,n_out],mean=0.5,stddev=0.5))    
}

biases = {
    'b1' : tf.Variable(tf.zeros([n_hidden1])),
    'b2' : tf.Variable(tf.zeros([n_hidden2])),
    'b3' : tf.Variable(tf.zeros([n_out]))
}
Y = tf.placeholder("float",[None,n_out])
```

```
#forward propagation 
def nn(X):
  layer_1 = tf.add(tf.matmul(X,weights['w1']),biases['b1'])
  layer_2 = tf.add(tf.matmul(layer_1,weights['w2']),biases['b2'])
  out_layer = tf.add(tf.matmul(layer_2,weights['w3']),biases['b3'])
  return out_layer

logits = nn(X)
```

```
#loss function
loss  = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=logits,labels=Y))
```

```
#backpropagation
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
train_ops = optimizer.minimize(loss)
```

```
#initializing variables
init = tf.global_variables_initializer()
```

```
#training
with tf.Session() as sess:
  sess.run(init)
  for epoch in range(epochs):
    x, y = mnist.train.next_batch(batch_size)
    sess.run(train_ops,feed_dict={X:x, Y:y})
    if ((epoch+1)%100) == 0:
      batch_loss=sess.run(loss,feed_dict={X:x, Y:y})
      print("Epoch : {} loss: {}".format(epoch+1,batch_loss))
  print("optimization done")  
  #testing
  correct_pred = tf.equal(tf.argmax(logits,1),tf.argmax(Y,1))
  accuracy = tf.reduce_mean(tf.cast(correct_pred,tf.float32))
  print("testing accuracy:",accuracy.eval({X: mnist.test.images,
                                        Y: mnist.test.labels}))
```
